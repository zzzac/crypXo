```python
import pandas as pd
from pathlib import Path
```

```python
data_dir = Path('data/binance/1m/ETH-USDT/')
parquet_files = sorted(data_dir.glob('*.parquet'))

df_list = []
for file in parquet_files:
    df_day = pd.read_parquet(file)
    df_list.append(df_day)
```
